```python
df['Events'] = df['Events'].replace(r'^\s*$', np.nan, regex=True)
```

```python
df['PrecipitationSumInches'] = df['PrecipitationSumInches'].replace('T', 0.0)
```

```python
df.info()
```

```python
numeric_cols = [
    'TempHighF', 'TempAvgF', 'TempLowF',
    'DewPointHighF', 'DewPointAvgF', 'DewPointLowF',
    'HumidityHighPercent', 'HumidityAvgPercent', 'HumidityLowPercent',
    'SeaLevelPressureHighInches', 'SeaLevelPressureAvgInches', 'SeaLevelPressureLowInches',
    'VisibilityHighMiles', 'VisibilityAvgMiles', 'VisibilityLowMiles',
    'WindHighMPH', 'WindAvgMPH', 'WindGustMPH',
    'PrecipitationSumInches'
]
```

```python
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
print(df.isnull().sum())
```

```python
df['Events'] = df['Events'].fillna('None')
```

```python
df.isnull().sum()
```

```python
df = df.sort_values('Date')
```

```python
df.head()
```

```python
cols_to_check = [
    'DewPointHighF', 'DewPointAvgF', 'DewPointLowF',
    'HumidityHighPercent', 'HumidityAvgPercent', 'HumidityLowPercent',
    'SeaLevelPressureHighInches', 'SeaLevelPressureAvgInches', 'SeaLevelPressureLowInches',
    'VisibilityHighMiles', 'VisibilityAvgMiles', 'VisibilityLowMiles',
    'WindHighMPH', 'WindAvgMPH', 'WindGustMPH'
]
```

```python
df[cols_to_check] = df[cols_to_check].interpolate(method='linear')
```

```python
table_id = "projecttest1-461318.etl_project_data.weather_data"

job_config = bigquery.LoadJobConfig(
    source_format=bigquery.SourceFormat.PARQUET,
)

with open("/content/temiz_data.parquet", "rb") as source_file:
    job = client.load_table_from_file(source_file, table_id, job_config=job_config)

job.result()  # yükləmə bitənə qədər gözlə
print("Yükləndi:", job.output_rows, "sətir")
```

```python
query = f"SELECT COUNT(*) as total_rows FROM `{table_id}`"
result = client.query(query).result()
for row in result:
    print(row.total_rows)
```
